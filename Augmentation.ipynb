```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# image read
img = cv2.imread('ISIC_0000000.jpg')

# BGR -> RGB
img = img[:,:,[2,1,0]]

plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

print(img.shape)
```

```python

for rotate in range(50):
    
    rnum = np.random.rand(3)
    # size range 50~100
    r_size = (rnum[0] * 50) + 50
    r_size = r_size.astype(np.int32)
    
    # row start index 300 ~ 900
    x_index = (rnum[2] * 600) + 300
    x_index = x_index.astype(np.int32)
    
    # col start index 20 ~ 667
    y_index = (rnum[1] * 647) + 20
    y_index = y_index.astype(np.int32)
    
    # crop_img = img[20:700, 50:900]

    crop_img = img[y_index:y_index+r_size, x_index:x_index+r_size]

    crop_img = cv2.resize(crop_img, (33,33), interpolation=cv2.INTER_LINEAR)

    cv2.imwrite("output/cropped_img"+str(rotate)+".jpg", crop_img[:,:,[2,1,0]])
    
    # rotate image
    rows, cols = crop_img.shape[0:2]

    M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
    #M2 = np.float32([[1,0,0],[0,1,-1]])
    rotated_img = cv2.warpAffine(crop_img, M, (cols,rows))
    #rotated_img = cv2.warpAffine(rotated_img, M2, (cols,rows))
    
    cv2.imwrite("output/rotated_img"+str(rotate)+".jpg", rotated_img[:,:,[2,1,0]])

    # V-flip image
    vflip_img = cv2.flip(crop_img,0)
    
    cv2.imwrite("output/vflip_img"+str(rotate)+".jpg", vflip_img[:,:,[2,1,0]])

    # H-flip image
    hflip_img = cv2.flip(crop_img,1)

    cv2.imwrite("output/hflip_img"+str(rotate)+".jpg", hflip_img[:,:,[2,1,0]])
    
    images = [crop_img, rotated_img, vflip_img, hflip_img]
    for i in range(4):
        plt.subplot(2,2,i+1), plt.imshow(images[i])
    plt.show()
```
