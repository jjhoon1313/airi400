```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('messi5.jpg',0)

print(img)

#cv2.imshow('test',img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()


# image show
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()

# image write
cv2.imwrite('messigray.png',img)

# draw sth(diagonal line)
img = np.zeros((512,512,3), np.uint8)

# img, startpoint, endpoint, color(RGB), thickness
cv2.line(img, (0,0), (511,511), (255,0,0), 5)

plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()


# draw circle
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(img, (x,y),100,(255,0,0),-1)
        
#create a black image, a window and bind the function to window
img = np.zeros([512,512,3], np.uint8)
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)

#while(1):
#    cv2.imshow('image',img)
#    if cv2.waitKey(20) & 0xFF == 27:
#        break
#    cv2.destroyAllWindows()


# pixel color change, get pixel attribute
img = cv2.imread('messi5.jpg')

plt.imshow(img[:,:,[2,1,0]], interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()

px = img[100,100,[2,1,0]]

blue = img[100,100,2]
print(blue)

img[100:111,100:111] = [255,255,255]

plt.imshow(img[:,:,[2,1,0]], interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()

# get attributes
print(img.item(10,10,2))
print(img.shape)
print img.size
print img.dtype


# cut & copy & paste image(numpy is better)
img = cv2.imread('messi5.jpg',0)

ball = img[125:245, 195:315]
img[280:400, 300:420] = ball

plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()

#split & merge images(numpy is better)
img = cv2.imread('messi5.jpg')
r,g,b = cv2.split(img)


img = cv2.merge((b,g,r))

plt.imshow(img, interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(img, interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()


# addWeighted image w, (1-w)
img1 = cv2.imread('ml.png')
img2 = cv2.imread('opencv_logo.jpg')

dst = cv2.addWeighted(img1, 0.8, img2, 0.2, 0)

plt.imshow(dst, interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()

# threshold
img = cv2.imread('messi5.jpg', 0)
ret, thresh1 = cv2.threshold(img, 127,255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 127,255, cv2.THRESH_BINARY_INV)

images = [thresh1, thresh2]
for i in range(2):
    plt.subplot(2,1,i+1), plt.imshow(images[i],cmap='gray')
plt.show()


# adaptive threshold
img2 = cv2.medianBlur(img,5)

ret, thresh3 = cv2.threshold(img2, 127,255,cv2.THRESH_BINARY)
thresh4 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 2)
thresh5 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 2)

images = [img2, thresh3, thresh4, thresh5]

for i in range(4):
    plt.subplot(2,2,i+1), plt.imshow(images[i], cmap='gray')
plt.show()


# resize image
img = cv2.imread('messi5.jpg')

res = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

plt.imshow(res[:,:,[2,1,0]], interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()

res.size

# move image
img = cv2.imread('messi5.jpg',0)

rows, cols = img.shape

M = np.float32([[1,0,500],[0,1,-200]])
dst = cv2.warpAffine(img,M,(cols,rows))

plt.imshow(dst, cmap='gray', interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()


# rotate images
M = cv2.getRotationMatrix2D((cols/2,rows/2),-55,1)
dst = cv2.warpAffine(img, M, (cols,rows))

plt.imshow(dst, cmap='gray', interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()

```
